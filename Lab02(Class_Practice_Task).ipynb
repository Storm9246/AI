```python
class SmartLights:
  house = "My house"
  def __init__(self, name):
    self.roomName = name
    self.status = False

  def checkStatus(self):
    if(self.status):
      print(f"{self.roomName} light is ON")
    else:
      print(f"{self.roomName} light is OFF")

  def switch(self):
    if(self.status):
      self.status = False
      print(f"{self.roomName} light turned OFF")
    else:
      self.status = True
      print(f"{self.roomName} light turned ON")


light1 = SmartLights("Bedroom")
light2 = SmartLights("Living Room")
light3 = SmartLights("Kitchen")

light1.switch()
light3.switch()
light1.checkStatus()
light2.checkStatus()
light3.checkStatus()
light3.switch()
light3.checkStatus()
```

```python
class Staff:
  def __init__(self, name, staff_ID, department):
    self.name = name
    self.staff_ID = staff_ID
    self.department = department

class teacher(Staff):
  def __init__(self, name, staff_ID, department, course, salary):
    super().__init__(name, staff_ID, department)
    self.course = course
    self.salary = salary

  def teach(self):
    print(f"{self.name} of department {self.department} have tought course {self.course}")

class Administrative_Staff(Staff):
  def __init__(self, name, staff_ID, department, role, work_Hours):
    super().__init__(name, staff_ID, department)
    self.role = role
    self.work_Hours = work_Hours

  def performTask(self):
    print(f"{self.name} have performed {self.role} for {self.work_Hours} hours")

class Research_Assistants(Staff):
  def __init__(self, name, staff_ID, department, research_topic, stipend):
    super().__init__(name, staff_ID, department)
    self.research_topic = research_topic
    self.stipend = stipend

  def research(self):
    print(f"{self.name} have researched in {self.research_topic}")


t1 = teacher("Dr. Atif Aslam", "T1001", "Computer Science", "Python Programming", 5000)
a1 = Administrative_Staff("Hania Amir", "A2005", "Registrar", "Student Enrollment", 40)
r1 = Research_Assistants("Jackie Chan", "R3002", "Physics", "Quantum Computing", 1500)

print("--- University Management System ---")

t1.teach()
a1.performTask()
r1.research()

print("\n--- Quick ID Check ---")
print(f"Staff Name: {t1.name} | ID: {t1.staff_ID}")
print(f"Staff Name: {a1.name} | ID: {a1.staff_ID}")
```
