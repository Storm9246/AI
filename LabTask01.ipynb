```python
name = input("Enter name: ")
marks = int(input("Enter marks: "))
if(marks >= 85 and marks <= 100):
  grade = 'A'
elif(marks >= 70):
  grade = 'B'
elif(marks >= 50):
  grade = 'C'
elif(marks < 50 and marks >= 0):
  grade = 'Fail'

print("Student name: ", name)
print("Marks: ", marks)
print("Grade: ", grade)
```

```python
n = int(input("Enter a number: "))
count = 0
even = []
for x in range(1, n+1):
  if(x % 2 == 0):
    even.append(x)
    count += 1
print("Even nummbers: ", even)
print("Total even numbers", count)
```

```python
record = {}
for x in range(3):
  record['name' + str(x)] = input("Enter student name: ")
  record['marks' + str(x)] = int(input("Enter marks: "))

for x in range(3):
  print(record['name' + str(x)], ' : ', record['marks' + str(x)])
```

```python
exit = True
while(exit):
  print("1. Add two numbers")
  print("2. Subtract two numbers")
  print("3. Exit")
  choice = int(input("Enter your choice: "))
  if(choice == 3):
    exit = False
    break
  elif(choice == 1):
    num1 = int(input("Enter first number: "))
    num2 = int(input("Enter second number: "))
    result = num1 + num2
    print("Result: ", result)
  elif(choice == 2):
    num1 = int(input("Enter first number: "))
    num2 = int(input("Enter second number: "))
    result = num1 - num2
    print("Result: ", result)
```

```python
def average(list):
  avg = 0
  for x in range(len(list)):
    avg += list[x]
  return avg / len(list)

marks = []
n = int(input("Enter quantity of marks: "))
for x in range(n):
  marks.append(int(input("Enter marks: ")))

print("Marks: ", marks)
print("Average: ", average(marks))
```
