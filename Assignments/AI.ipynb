```python
class MazeSolver:
    def __init__(self, grid):
        self.grid = grid
        self.height = len(grid)
        self.width = len(grid[0])
        self.start_node = self._locate_symbol('S')
        self.goal_node = self._locate_symbol('G')

    def _locate_symbol(self, target):
        """Finds coordinates of a specific character in the grid."""
        for r, row in enumerate(self.grid):
            for c, val in enumerate(row):
                if val == target:
                    return (r, c)
        return None

    def _is_accessible(self, r, c):
        """Checks if a cell is within bounds and not a wall."""
        return 0 <= r < self.height and 0 <= c < self.width and self.grid[r][c] != 1

    def _run_dls(self, current, limit, path, visited):
        """Core Recursive Depth-Limited Search."""
        if current == self.goal_node:
            return path

        if limit <= 0:
            return None

        r, c = current
        # Directional moves: Down, Up, Right, Left
        for dr, dc in [(1, 0), (-1, 0), (0, 1), (0, -1)]:
            neighbor = (r + dr, c + dc)

            if self._is_accessible(*neighbor) and neighbor not in visited:
                visited.add(neighbor)
                result = self._run_dls(neighbor, limit - 1, path + [neighbor], visited)
                
                if result:
                    return result
                
                # Backtrack: allow other branches to explore this node if needed
                visited.remove(neighbor)

        return None

    def solve_dls(self, depth_limit):
        """Public interface for DLS."""
        if not self.start_node: return None
        return self._run_dls(self.start_node, depth_limit, [self.start_node], {self.start_node})

    def solve_ids(self, max_search_depth):
        """Public interface for Iterative Deepening."""
        for depth in range(max_search_depth + 1):
            found_path = self.solve_dls(depth)
            if found_path:
                return found_path
        return None

# --- Execution ---

maze_data = [
    ['S', 0, 0, 1, 0],
    [1, 0, 1, 0, 0],
    [0, 0, 0, 0, 'G'],
    [1, 1, 0, 1, 1]
]

solver = MazeSolver(maze_data)

print("--- Depth Limited Search (Limit: 6) ---")
print(solver.solve_dls(6))

print("\n--- Iterative Deepening Search ---")
print(solver.solve_ids(10))
```
