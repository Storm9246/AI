```python
class TrfcEnv:
    def __init__(self, st):
        self.st = st

    def get_st(self):
        return self.st

class RefAgnt:
    def act(self, p):
        if p == "Heavy Traffic":
            return "Extend Green Time"
        else:
            return "Normal Green"

bot = RefAgnt()

s1 = TrfcEnv("Heavy Traffic")
p1 = s1.get_st()
act1 = bot.act(p1)
print(f"Percept: {p1} -> Action: {act1}")

s2 = TrfcEnv("Light Traffic")
p2 = s2.get_st()
act2 = bot.act(p2)
print(f"Percept: {p2} -> Action: {act2}")
```

```python
import random

class ClsRoom:
    def __init__(self):
        self.ppl = random.choice(["Yes", "No"])
        self.bulb = "OFF"

    def sense(self):
        return {'ppl': self.ppl, 'lgt': self.bulb}

    def chg_ppl(self):
        self.ppl = random.choice(["Yes", "No"])

    def on(self):
        self.bulb = "ON"

    def off(self):
        self.bulb = "OFF"

class MdlAgent:
    def __init__(self):
        self.mem = {'p_ppl': None, 'p_lgt': None, 'c_ppl': None, 'c_lgt': None}

    def do_job(self, data):
        self.mem['p_ppl'] = self.mem['c_ppl']
        self.mem['p_lgt'] = self.mem['c_lgt']
        self.mem['c_ppl'] = data['ppl']
        self.mem['c_lgt'] = data['lgt']

        if self.mem['c_ppl'] == "Yes" and self.mem['c_lgt'] == "OFF":
            return "Sw On"
        elif self.mem['c_ppl'] == "No" and self.mem['c_lgt'] == "ON":
            return "Sw Off"
        else:
            return "Idle"

rm = ClsRoom()
bot = MdlAgent()

for x in range(8):
    print(f"\nStep {x + 1}:")

    rm.chg_ppl()
    p = rm.sense()
    act = bot.do_job(p)

    if act == "Sw On":
        rm.on()
    elif act == "Sw Off":
        rm.off()

    print(f"Seen: {p}")
    print(f"ToDo: {act}")
    print(f"Mem: {bot.mem}")
```

```python
class GoalAgnt:
    def __init__(self, t_list):
        self.t_list = t_list
        self.done = []

    def start_work(self):
        for t in self.t_list:
            if t not in self.done:
                print(f"Studying {t}")
                self.done.append(t)

        if len(self.done) == len(self.t_list):
            print("Goal Achieved: All subjects completed")

subs = ["AI", "Math", "Physics"]
bot = GoalAgnt(subs)
bot.start_work()
```

```python
class UtlAgnt:
    def __init__(self, opts):
        self.opts = opts

    def choose(self):
        res = {}
        for n, v in self.opts.items():
            val = v['R'] - v['D']
            res[n] = val
            print(f"Rest {n} Util: {val}")

        top = max(res, key=res.get)
        print(f"Choice: {top}")

data = {
    'A': {'D': 3, 'R': 7},
    'B': {'D': 5, 'R': 9}
}

bot = UtlAgnt(data)
bot.choose()
```

```python
import random

class LrnBot:
    def __init__(self):
        self.q_vals = {'Play': 0, 'Rest': 0}
        self.lr = 0.1
        self.rwds = {'Play': 5, 'Rest': 1}

    def learn(self, act):
        cur = self.q_vals[act]
        r = self.rwds[act]
        self.q_vals[act] = cur + self.lr * (r - cur)

agent = LrnBot()

for i in range(10):
    a = random.choice(list(agent.rwds.keys()))
    rew = agent.rwds[a]
    agent.learn(a)
    print(f"Step {i+1}: Action {a} Reward {rew}")

print("Q-table Updated")
print(agent.q_vals)
```

```python
class BldgEnv:
    def __init__(self):
        self.map = {
            "a": "OK", "b": "OK", "c": "FIRE",
            "d": "OK", "e": "FIRE", "f": "OK",
            "g": "OK", "h": "OK", "j": "FIRE"
        }
        self.route = ["a", "b", "c", "d", "e", "f", "g", "h", "j"]

    def scan(self, loc):
        return self.map[loc]

    def clear(self, loc):
        self.map[loc] = "OK"

    def show(self):
        print("Status:")
        cnt = 0
        for r in self.route:
            if self.map[r] == "OK":
                print("[ ]", end=" ")
            else:
                print("[*]", end=" ")

            cnt += 1
            if cnt % 3 == 0:
                print()
        print()

class FireBot:
    def __init__(self):
        self.path = ["a", "b", "c", "d", "e", "f", "g", "h", "j"]

    def patrol(self, area):
        for rm in self.path:
            print(f"Mov to {rm}...")
            st = area.scan(rm)
            if st == "OK":
                print(f"{rm} is safe.")
            else:
                print(f"Fire at {rm}! Extinguishing...")
                area.clear(rm)
                print(f"{rm} cleared.")

            area.show()
        print("Patrol Complete.")

area = BldgEnv()
bot = FireBot()
bot.patrol(area)

print("Final View:")
area.show()
```
